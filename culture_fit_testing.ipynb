```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    "Job": ["Job 1", "Job 2", "Job 3", "Job 4", "Job 5", "Job 6", "Job 7"],
    "BERT Score": [0.40, 0.31, 0.341, 0.404, 0.2561, 0.307, 0.324],
    "Gemini Score": [5, 3, 3, 4, 4, 4, 5]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert BERT scores to percentages
df["BERT Accuracy (%)"] = (df["BERT Score"] * 10).round(2)

# Plotting
plt.figure(figsize=(10, 6))

# Bar chart for BERT Accuracy and Gemini Scores
plt.bar(df["Job"], df["BERT Accuracy (%)"], color="blue", alpha=0.7, label="BERT Accuracy (%)")
plt.bar(df["Job"], df["Gemini Score"], color="orange", alpha=0.7, label="Gemini Score", width=0.4)

# Adding details
plt.title("Comparison of BERT Accuracy (%) and Gemini Scores", fontsize=14)
plt.xlabel("Job", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.legend(loc="upper right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()

# Print the updated DataFrame for reference
print(df)
```
